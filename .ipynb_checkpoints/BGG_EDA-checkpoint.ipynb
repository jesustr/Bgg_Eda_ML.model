###  What is avg_rating and geek_rating ? 

avg_rating = The average of the votes cast by people who participated in the vote.(num_votes)


geek_rating = Bayesian averaging the votes cast by people who participated in the vote.(num_votes) 


age = age limit (13 = 13+)

#### What is Bayesian Averaging ?

The Bayesian average adjusts the average rating of products whose rating counts fall below a threshold. Suppose the threshold amount is calculated to be 100. That means average ratings with less than 100 ratings get adjusted, while average ratings with more than 100 ratings change only very slightly

# ML MODEL 

```python
end_games.head()
```

## X-y , Train-Test 

```python
X = end_games.drop(['rank','name','game_id','owned','avg_rating',\
                   'geek_rating','num_votes','owned','designer'],axis = 1 ).astype('float')
```

```python
y = end_games['avg_rating']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size = 0.20,
                                                    random_state = 34)
```

# XGBoost / Model 

```python
xgb_model = XGBRegressor().fit(X_train,y_train)
```

### Predict 

```python
# primitive test error
y_pred = xgb_model.predict(X_test)
np.sqrt(mean_squared_error(y_test,y_pred))
```

## MODEL TUNING 

```python
xgb_model
```

```python
xgb_grid = {
    'colsample_bytree': [0.6,0.9,1],
    'n_estimators': [100,200,500,1000],
    'max_depth': [2,3,4,5,6],
    'learning_rate': [0.1,0.01,0.5]
}
```

```python
xgb = XGBRegressor()

xgb_cv = GridSearchCV(xgb,
                      param_grid = xgb_grid,
                      cv = 10,
                      n_jobs = -1, 
                      verbose = 2)

xgb_cv.fit(X_train, y_train)
```

```python
xgb_cv.best_params_
```

### We refitted the data using the tuned model. 

```python
xgb_tuned = XGBRegressor(colsample_bytree = 0.6,
                         learning_rate = 0.1,
                         max_depth = 6,
                         n_estimators = 500)

xgb_tuned = xgb_tuned.fit(X_train,y_train)
```

### Our tuned mean squared error. 

```python
y_pred = xgb_tuned.predict(X_test)
```

```python
print(f'Tuned mean squared eror = {np.sqrt(mean_squared_error(y_test, y_pred))}' )
```

### If we set all our predictions as the average value, what would our mean squared error be?

```python
y_predmean = y_pred.copy()
```

```python
y_predmean.fill(y_test.mean())
```

```python
np.sqrt(mean_squared_error(y_test, y_predmean))
```

```python
# If we estimated the average value for each row, we would get a result of 0.43292 MSE.
# Our predict models result 0.28721 MSE.
```
